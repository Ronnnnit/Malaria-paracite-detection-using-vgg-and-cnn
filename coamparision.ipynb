```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, roc_curve, auc, precision_recall_curve

# Generate sample data for demonstration
y_true = np.random.randint(0, 2, size=100)
y_pred_cnn = np.random.rand(100)
y_pred_vgg16 = np.random.rand(100)

# Calculate confusion matrices
cm_cnn = confusion_matrix(y_true, (y_pred_cnn > 0.5).astype(int))
cm_vgg16 = confusion_matrix(y_true, (y_pred_vgg16 > 0.5).astype(int))

# Calculate ROC curve for CNN
fpr_cnn, tpr_cnn, _ = roc_curve(y_true, y_pred_cnn)
roc_auc_cnn = auc(fpr_cnn, tpr_cnn)

# Calculate ROC curve for VGG16
fpr_vgg16, tpr_vgg16, _ = roc_curve(y_true, y_pred_vgg16)
roc_auc_vgg16 = auc(fpr_vgg16, tpr_vgg16)

# Calculate precision-recall curve for CNN
precision_cnn, recall_cnn, _ = precision_recall_curve(y_true, y_pred_cnn)

# Calculate precision-recall curve for VGG16
precision_vgg16, recall_vgg16, _ = precision_recall_curve(y_true, y_pred_vgg16)

# Print data for CNN
print("CNN:")
print("Confusion Matrix:")
print(cm_cnn)
print("ROC AUC:", roc_auc_cnn)
print()

# Print data for VGG16
print("VGG16:")
print("Confusion Matrix:")
print(cm_vgg16)
print("ROC AUC:", roc_auc_vgg16)
print()

# Plotting
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# CNN
""" axes[0, 0].imshow(cm_cnn, cmap=plt.cm.Blues)
axes[0, 0].set_title('Confusion Matrix (CNN)')
axes[0, 0].set_xlabel('Predicted label')
axes[0, 0].set_ylabel('True label') """

axes[0, 0].plot(fpr_cnn, tpr_cnn, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc_cnn)
axes[0, 0].plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
axes[0, 0].set_xlim([0.0, 1.0])
axes[0, 0].set_ylim([0.0, 1.05])
axes[0, 0].set_xlabel('False Positive Rate')
axes[0, 0].set_ylabel('True Positive Rate')
axes[0, 0].set_title('Receiver Operating Characteristic (CNN)')
axes[0, 0].legend(loc="lower right")

""" axes[1, 0].step(recall_cnn, precision_cnn, color='b', alpha=0.2, where='post')
axes[1, 0].fill_between(recall_cnn, precision_cnn, alpha=0.2, color='b', step='post')
axes[1, 0].set_xlabel('Recall')
axes[1, 0].set_ylabel('Precision')
axes[1, 0].set_ylim([0.0, 1.05])
axes[1, 0].set_xlim([0.0, 1.0])
axes[1, 0].set_title('Precision-Recall curve (CNN)') """
# VGG16
""" axes[1, 1].imshow(cm_vgg16, cmap=plt.cm.Blues)
axes[1, 1].set_title('Confusion Matrix (VGG16)')
axes[1, 1].set_xlabel('Predicted label')
axes[1, 1].set_ylabel('True label') """

axes[0, 1].plot(fpr_vgg16, tpr_vgg16, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc_vgg16)
axes[0, 1].plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
axes[0, 1].set_xlim([0.0, 1.0])
axes[0, 1].set_ylim([0.0, 1.05])
axes[0, 1].set_xlabel('False Positive Rate')
axes[0, 1].set_ylabel('True Positive Rate')
axes[0, 1].set_title('Receiver Operating Characteristic (VGG16)')
axes[0, 1].legend(loc="lower right")

""" axes[1, 1].step(recall_vgg16, precision_vgg16, color='b', alpha=0.2, where='post')
axes[1, 1].fill_between(recall_vgg16, precision_vgg16, alpha=0.2, color='b', step='post')
axes[1, 1].set_xlabel('Recall')
axes[1, 1].set_ylabel('Precision')
axes[1, 1].set_ylim([0.0, 1.05])
axes[1, 1].set_xlim([0.0, 1.0])
axes[1, 1].set_title('Precision-Recall curve (VGG16)') """

plt.tight_layout()
plt.show()


```
